### Вот краткий обзор различных моделей, используемых при обнаружении объектов.

#### Семейство моделей R-CNN включает:

1. R-CNN — использует метод выборочного поиска для обнаружения областей интереса на входных изображениях и использует классификатор на основе DCN (глубокой сверточной нейронной сети) для независимой классификации областей интереса.

2. SPPNet и Fast R-CNN — это улучшенные версии R-CNN, которые занимаются извлечением областей интереса из карт признаков (feature maps). Они намного быстрее, чем обычная архитектура R-CNN.

3. Faster R-CNN — это улучшенная версия Fast R-CNN, которая была обучена от начала до конца путем введения RPN (region proposal network - сеть региональных предложений). RPN — это сеть, используемая для создания области интереса путем регрессии блоков привязки (anchor boxes). Следовательно, блоки привязки затем используются в задаче обнаружения объектов.

4. Mask R-CNN добавляет ветвь предсказания маски в Faster R-CNN, которая может одновременно обнаруживать объекты и предсказывать их маски.

5. R-FCN заменяет полносвязные слои чувствительными к положению картами оценок для лучшего обнаружения объектов.

6. Cascade R-CNN решает проблему переобучения при обучении и несоответствия качества при выводе путем обучения последовательности детекторов с увеличением порогов IoU.


#### Семейство моделей YOLO включает:

1. YOLO использует меньше кол-во блоков привязки (разделяет входное изображение на сетку S × S) для регрессии и классификации. Это было построено с использованием нейронных сетей darknet.

2. YOLOv2 повышает производительность за счет использования большего количества блоков привязки и нового метода регрессии ограничивающей рамки.

3. YOLOv3 — это расширенная версия варианта v2 с более глубокой сетью детекторов признаков и незначительными изменениями представления.
YOLOv3 имеет относительно быстрое время вывода, которое занимает примерно 30 мс на вывод.

4. YOLOv4 (обновление YOLOv3) работает путем разбиения задачи обнаружения объекта на две части, регрессии для определения позиционирования объекта с помощью ограничительных рамок и классификации для определения класса объекта. YOLO V4 и его преемники технически являются продуктом другой группы исследователей, чем версии 1-3.

5. YOLOv5 — это улучшенная версия YOLOv4 с методом мозаичного увеличения для повышения общей производительности YOLOv4.

#### Семейство моделей CenterNet включает:

1. SSD плотно размещает блоки привязки поверх входного изображения и использует функции из разных сверточных слоев для регрессии и классификации блоков привязки.

2. DSSD представляет модуль деконволюции в SSD для объединения низкоуровневых и высокоуровневых функций. В то время как R-SSD использует операции объединения и деконволюции в разных слоях объектов, чтобы объединить низкоуровневые и высокоуровневые функции.

3. RON предлагает обратную связь и объектность до эффективного извлечения многомасштабных признаков.

4. RefineDet уточняет расположение и размеры блоков привязки в 2 раза, что наследует достоинства как одноэтапного, так и двухэтапного подходов.

5. CornerNet — это еще один подход, основанный на ключевых точках, который напрямую обнаруживает объект, используя пару углов. Хотя CornerNet достигает высокой производительности, ему еще есть куда расти.

6. CenterNet исследует визуальные шаблоны внутри каждого ограничивающего блока. Для обнаружения объекта используется 3, а не 2 ключевых точек.

#### CenterNet оценивает объекты как отдельные точки, прогнозируя координаты x и y центра объекта и его зону охвата (ширину и высоту). Это уникальная технология, которая превосходит такие варианты, как SSD и семейство R-CNN.
